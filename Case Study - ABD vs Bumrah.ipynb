# Case Study : AB de Villiers vs Jasprit Bumrah ( IPL Face-off )

```python
# importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# to display all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', None)
```

```python
# ipl-ball-by-ball-dataset
df = pd.read_csv("IPL_ball_by_ball_updated.csv")
```

```python
# taking a look at the data we got

# top-5 rows
df.head()
```

```python
# bottom-5 rows
df.tail()
```

We can see that this dataset contains data from :
 - Season - 2008 to Season - 2021
 - Ball-by-ball data is available

```python
# Time to explore the data

# checking the data for number of innings
df.innings.unique()
```

We can see that it has total six innings data.
 - Innings 3 and 4 are for the super-over
 - Innings 5 and 6 are for double super-over

But we need only data for Innings 1 and 2. So we have to filter out based on innings.

```python
# selecting only innings 1 and 2
df = df[(df.innings==1) | (df.innings==2)]
```

```python
df.innings.unique()
```

- Now it's time to move towards filtering data based on player choics.
- For that we have to find the exact name of required players that is present in the data.

```python
# checking the team names
df.batting_team.unique()
```

Required teams are :
   - Royal Challengers Bangalore
   - Mumbai Indians

```python
# searching our striker name
df[df.batting_team == 'Royal Challengers Bangalore']['striker'].unique()
```

```python
# searching our bowler name
df[df.bowling_team == 'Mumbai Indians'].bowler.unique()
```

```python
# Player 1 : AB de Villiers (striker)
# Player 2 : JJ Bumrah (bowler)
```

Now the main filtering need to be done with our striker and bowler only in our data

```python
# filtering based on Player 1 and Player 2
req_df = df[(df.striker == 'AB de Villiers') & (df.bowler == 'JJ Bumrah')]

req_df.head()
```

```python
req_df.tail()
```

So, now we are done with the data filtering part.


```python
# data we are looking for
# What are the numbers when ABD faces Bumrah ?
# No. of runs sccored ?
# No. of balls faced ?
# No. of times dismissed ?
```

```python
# No. of runs scored by ABD against Bumrah
sum(req_df.runs_off_bat)
```

```python
# No. of balls faced
len(req_df)
```

```python
# No. of dismissals
len(req_df[req_df.player_dismissed == "AB de Villiers"])
```

```python
# ABD's strike rate against Bumrah
100*sum(req_df.runs_off_bat)/len(req_df)
```

```python
# 6's and 4's by ABD

print("6's : ",len(req_df[req_df['runs_off_bat'] == 6]), " - 4's : ",len(req_df[req_df['runs_off_bat'] == 6]))
```

```python
# Bumrah's economy against ABD
6*sum(req_df.runs_off_bat)/len(req_df)
```

```python
# Step 1 : Filter by player names - Done
# Step 2 : Use these names and assign it to striker and bowler - Done
# Step 3 : Get the required data - Done
```

```python
# bumrah data-frame
bumrah_df = df[df.bowler == 'JJ Bumrah']
bumrah_df.head(3)
```

```python
# runs scored by batsman against bumrah
bdf1 = pd.DataFrame(bumrah_df.groupby("striker")['runs_off_bat'].sum()).reset_index()
bdf1.head(3)
```

```python
# balls faced by batsman
bdf2 = pd.DataFrame(bumrah_df.groupby("striker")["ball"].count()).reset_index()
bdf2.head(2)
```

```python
# merging the two dataframes
bdf3 = bdf1.merge(bdf2, on="striker", how="left")
bdf3.head(3)
```

```python
# adding strike-rate column
bdf3["strike_rate"] = 100*bdf3["runs_off_bat"]/bdf3["ball"]
bdf3.head(3)
```

```python
# here we will choose only those strikers who faced minimum 30 balls
# min-criteria : 30 balls
bdf3 = bdf3[bdf3.ball >= 30]
bdf3.head(3)
```

```python
# ABD data-frame
abd_df = df[df.striker == 'AB de Villiers']
abd_df.head(3)
```

```python
# runs scored by ABD against different bowlers
adf1 = pd.DataFrame(abd_df.groupby("bowler")['runs_off_bat'].sum()).reset_index()
adf1.head(3)
```

```python
# balls faced by ABD 
adf2 = pd.DataFrame(abd_df.groupby("bowler")['ball'].count()).reset_index()
adf2.head(3)
```

```python
adf3 = adf1.merge(adf2, on="bowler", how="left")
adf3.head(3)
```

```python
adf3["strike_rate"] = 100*adf3["runs_off_bat"]/adf3["ball"]
adf3.head(3)
```

```python
# min-criteria : 30 balls
adf3 = adf3[adf3.ball >= 30]
adf3.head(3)
```

```python
# bdf3
bdf3.reset_index(inplace=True, drop=True)

# adf3
adf3.reset_index(inplace=True, drop=True)
```

- Visualizing insights

```python
# Lets try some visualizations to see the clear picture
```

```python
# plotting the batsman who faced Bumrah in IPL (min 30 balls)

plt.figure(figsize = (16,8))
plt.scatter(bdf3.strike_rate, bdf3.runs_off_bat)

for i in range(len(bdf3)):
    if(bdf3['striker'][i] == "V Kohli"):
        plt.text(bdf3["strike_rate"][i] + 1.5, bdf3["runs_off_bat"][i], bdf3["striker"][i])
    elif(bdf3['striker'][i] == "AB de Villiers"):
        plt.text(bdf3["strike_rate"][i], bdf3["runs_off_bat"][i] - 4, bdf3["striker"][i], backgroundcolor = 'lightgreen')
    elif(bdf3["strike_rate"][i] >= 120 and bdf3["runs_off_bat"][i] >= 60):
        plt.text(bdf3["strike_rate"][i], bdf3["runs_off_bat"][i] - 3.5, bdf3["striker"][i])    

plt.axvline(120, ls="--", color="grey")
plt.axhline(60, ls="--", color="grey")
plt.title("Best Batsman against Bumrah in IPL (min 30 balls faced)", fontsize = 20)
plt.xlabel("Strike rate")
plt.ylabel("Runs scored")
plt.show()
```

##### Key Insight :
- ABD is among the batsmans who threatened Bumrah

```python
# plotting bowlers whom ABD faced (min 30 balls)

plt.figure(figsize = (16, 8))

plt.scatter(adf3["strike_rate"], adf3["runs_off_bat"])

for i in range(len(adf3)):
    if(adf3["bowler"][i] == "AD Russell"):
        plt.text(adf3["strike_rate"][i], adf3["runs_off_bat"][i] - 3.5, adf3["bowler"][i])
    elif(adf3["bowler"][i] == "JJ Bumrah"):
        plt.text(adf3["strike_rate"][i] + 2, adf3["runs_off_bat"][i], adf3["bowler"][i], backgroundcolor='lightblue')
    elif(adf3["strike_rate"][i] >= 130 and adf3["runs_off_bat"][i] >= 70):
        plt.text(adf3["strike_rate"][i] + 1.5, adf3["runs_off_bat"][i], adf3["bowler"][i])

plt.axvline(130, ls="--", color="grey")
plt.axhline(70, ls="--", color="grey")
plt.title("ABD's Favourite Bowlers (in IPL)", fontsize=20)
plt.xlabel("Strike rate")
plt.ylabel("Runs scored")
plt.show()
```

##### Key Insight :
- Bumrah is among the favourite bowlers of ABD.
